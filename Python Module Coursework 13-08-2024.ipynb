```python
# Define the abstract base class Person
class Person(ABC):
    def __init__(self, name, age, gender, address):
        self.name = name
        self.age = age
        self.gender = gender
        self.address = address

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}, Gender: {self.gender}, Address: {self.address}"

    def greet(self, other_person):
        print(f"Hello {other_person.name}! My name is {self.name}.")

    @abstractmethod
    def introduce(self):
        pass

    @staticmethod
    def is_adult(age):
        return age >= 18


# Define the Employee class that inherits from Person
class Employee(Person):
    # Class attribute to keep track of the number of instances
    _counter = 0

    def __init__(self, name, age, gender, address, salary):
        # Call the constructor of the Person class
        super().__init__(name, age, gender, address)
        # Increment counter and set employee_id
        Employee._counter += 1
        self._employee_id = f"EMP{Employee._counter:02d}"
        # Initialize protected salary attribute
        self._salary = salary

    # Getter for employee_id (private attribute)
    @property
    def employee_id(self):
        return self._employee_id

    # Property for the class attribute counter
    @property
    def counter(cls):
        return cls._counter
    counter = classmethod(counter)

    # Getter and setter for salary
    @property
    def salary(self):
        return self._salary

    @salary.setter
    def salary(self, value):
        if value < 0:
            self._salary = value

    # Method to increase salary
    def increase_salary(self, amount):
        if amount < 0:
            raise ValueError("Increase amount cannot be negative.")
        self._salary += amount

    # Method to decrease salary
    def decrease_salary(self, amount):
        if amount < 0:
            raise ValueError("Decrease amount cannot be negative.")
        if self._salary - amount < 0:
            raise ValueError("Salary cannot be negative.")
        self._salary -= amount

    # Override the abstract method introduce
    def introduce(self):
        return f"I am an employee. My name is {self.name} and my employee ID is {self._employee_id}."

    # Destructor to decrement counter when an instance is deleted
    def __del__(self):
        Employee._counter -= 1
```
