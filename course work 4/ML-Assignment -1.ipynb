### 1.)  Write a function to generate an m+1 dimensional data set, of size n, consisting of m continuous independent 
### 
variables (X) and one dependent variable (Y) defined a  yi = xiβ + es           e xi0 = 1

```python
# a function to generate an m+1 dimensional data set
def dataset(sigma, n, m):
    # Here sigma is the Standard deviation of the Gaussian noise
    # n is int type Number of samples
    # m is the Number of continuous independent variables
    
    # Generate random coefficients
    beta = np.random.rand(m + 1)
    # Generate random values for the independent variables
    X = np.hstack((np.ones((n, 1)), np.random.rand(n, m)))
    # Generate random noise e from a Gaussian distribution
    e = np.random.normal(0, sigma, n)

    
    # Compute the dependent variable Y using the formula: Y = Xβ + e
    Y = X @ beta + e  # Matrix multiplication for Xβ
    return X, Y, beta



# Example usage:
sigma = 0.3  # Standard deviation of Gaussian noise
n = 200      # Number of samples
m = 5         # Number of independent variables
X, Y, beta = dataset(sigma, n, m)

print("Independent Variables (X):\n", X)
print("Dependent Variable (Y):\n", Y)
print("Coefficients (β):\n", beta)
    
```
